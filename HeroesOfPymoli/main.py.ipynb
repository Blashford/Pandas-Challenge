```python
import pandas as pd

inpath = "Resources/purchase_data.csv"

hero_data = pd.read_csv(inpath)
hero_data.head()
```

```python
total_players = hero_data["SN"].nunique()
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

```python
items = hero_data["Item Name"].nunique()
avg_price = hero_data["Price"].mean()
total_purchase = hero_data["Item Name"].count()
total_rev = hero_data["Price"].sum()

purchase_analysis_df = pd.DataFrame({
    "Number of Unique Items": [items],
    "Average Price": avg_price,
    "Number of Purchases": total_purchase,
    "Total Revenue": total_rev
})
purchase_analysis_df
```

```python
count = hero_data.groupby("Gender")["SN"].nunique()
genpercent = (count/total_players)*100
gender_df = pd.DataFrame({
    "Total Count": count,
    "Percentage of Players": genpercent
})
gender_df = gender_df.sort_values("Total Count", ascending=False)
gender_df_tidy = gender_df.copy()
gender_df_tidy.index.name = None
gender_df_tidy["Percentage of Players"] = gender_df_tidy["Percentage of Players"].map("{:.2f}%".format)
gender_df_tidy
```

```python
purchase_count_gen = hero_data.groupby("Gender")["SN"].count()
avg_price_gen = hero_data.groupby("Gender")["Price"].mean()
total_purch_gen = hero_data.groupby("Gender")["Price"].sum()
avg_per_person = total_purch_gen/count
gender_purchase_df = pd.DataFrame({
    "Purchase Count": purchase_count_gen,
    "Average Purchase Price": avg_price_gen,
    "Total Purchase Value": total_purch_gen,
    "Avg Total Purchase per Person": avg_per_person
})
gender_purchase_tidy = gender_purchase_df.copy()
gender_purchase_tidy["Average Purchase Price"] =  gender_purchase_tidy["Average Purchase Price"].map("${:.2f}".format)
gender_purchase_tidy["Total Purchase Value"] =  gender_purchase_tidy["Total Purchase Value"].map("${:.2f}".format)
gender_purchase_tidy["Avg Total Purchase per Person"] =  gender_purchase_tidy["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_purchase_tidy
```

```python
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 400000]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_data = hero_data.copy()
age_data["Age bin"] = pd.cut(age_data["Age"], age_bins, labels = age_labels, include_lowest = True)

age_group = age_data.groupby("Age bin")
age_count = age_group["SN"].nunique()

age_percent = (age_count/total_players)*100
age_count_df = pd.DataFrame({
    "Total Count": age_count,
    "Percentage of Players": age_percent
})


age_count_tidy = age_count_df.copy()
age_count_tidy.index.name = None
age_count_tidy["Percentage of Players"] = age_count_tidy["Percentage of Players"].map("{:.2f}%".format)
age_count_tidy
```

```python
age_purchase = age_group["Item Name"].count()
avg_price_age = age_group["Price"].mean()
total_purch_age = age_group["Price"].sum()
avg_per_person_age = total_purch_age/age_count

age_purchase_df = pd.DataFrame({
     "Purchase Count": age_purchase,
    "Average Purchase Price": avg_price_age,
    "Total Purchase Value": total_purch_age,
    "Avg Total Purchase per Person": avg_per_person_age
})
age_purchase_tidy = age_purchase_df.copy()
age_purchase_tidy["Average Purchase Price"] =  age_purchase_tidy["Average Purchase Price"].map("${:.2f}".format)
age_purchase_tidy["Total Purchase Value"] =  age_purchase_tidy["Total Purchase Value"].map("${:.2f}".format)
age_purchase_tidy["Avg Total Purchase per Person"] =  age_purchase_tidy["Avg Total Purchase per Person"].map("${:.2f}".format)
age_purchase_tidy.index.name = None
age_purchase_tidy
```

```python
spender_count = hero_data.groupby("SN")["Item Name"].count()
spend_price = hero_data.groupby("SN")["Price"].mean()
spend_total = hero_data.groupby("SN")["Price"].sum()

spend_df = pd.DataFrame({
    "Purchase Count": spender_count,
    "Average Purchase Price": spend_price,
    "Total Purchase Value": spend_total
})
spend_df = spend_df.sort_values("Purchase Count", ascending = False)

spend_tidy = spend_df.copy()
spend_tidy["Average Purchase Price"] = spend_tidy["Average Purchase Price"].map("${:.2f}".format)
spend_tidy["Total Purchase Value"] = spend_tidy["Total Purchase Value"].map("${:.2f}".format)
spend_tidy.head()
```

```python
item_prof = hero_data.groupby(["Item ID", "Item Name"])

item_prof_count = item_prof["Purchase ID"].count()
item_cost = item_prof["Price"].mean()
item_total = item_prof["Price"].sum()


item_df = pd.DataFrame({
    "Purchase Count": item_prof_count,
    "Item Price": item_cost,
    "Total Purchase Value": item_total
})
item_df = item_df.sort_values("Purchase Count", ascending = False)

item_tidy = item_df.copy()

item_tidy["Item Price"] = item_tidy["Item Price"].map("${:.2f}".format)
item_tidy["Total Purchase Value"] = item_tidy["Total Purchase Value"].map("${:.2f}".format)
item_tidy.head()
```

```python
item_price_df = item_df.copy()
item_price_df = item_price_df.sort_values("Total Purchase Value", ascending = False)

item_price_tidy = item_price_df.copy()

item_price_tidy["Item Price"] = item_price_tidy["Item Price"].map("${:.2f}".format)
item_price_tidy["Total Purchase Value"] = item_price_tidy["Total Purchase Value"].map("${:.2f}".format)

item_price_tidy.head()
```
